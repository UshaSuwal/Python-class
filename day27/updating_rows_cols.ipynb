```python
people= {"first":["Jon","Jane","Hary","Ken"],
         "last":["Suwal","Thapa","Shrestha","Lawaju"],
         "email":["jon@email.com","jane@email.com","hary@email.com","ken@email.com"]}
```

```python
import pandas as pd
df=pd.DataFrame(people)
df
```

```python
df.columns=["first_name","last_name","Email"]    # for all column name change
df
```

```python
df.columns=[x.upper() for x in df.columns]
df
```

```python
df.columns=df.columns.str.replace('_'," ")   # replacing '_' with space
df
```

```python
df.columns=[x.lower() for x in df.columns]
df
```

```python
df.columns=df.columns.str.replace(' ',"_")   # replacing '_' with space
df
```

```python
df.rename(columns={"email":"Email"}, inplace=True)   # inplace narakhey dataframe ma change hudaina. df matra print garda change dekhidaina
```

```python
df
```

```python
# Changing in Rows. Rows change garna we use loc or iloc because for rows, indexing is necessary
df.loc[2]=["hary","potter","hp@email.com"]   # for all element change
df
```

```python
df.loc[2,["first_name","last_name"]]=["Hary","Potter"]
df
```

```python
df.loc[2,"Email"]="hary@email.com"
df
```

```python
# Using Filter
filt=(df["Email"]=="hary@email.com")
df.loc[filt,"last_name"]="potter"
df
```

```python
# Updating Series / Column
def to_upper(email):
    return email.upper()
df["Email"]=df["Email"].apply(to_upper)     # sabai ekai patak change garna lai apply() use garne
df
```

```python
df["Email"]=df["Email"].apply(lambda x: x.lower())   # using lambda function
df
```

```python
# euta euta data change garna chai replace() use garne
df["first_name"]=df["first_name"].replace({"Jon":"John","Harry":"Hary"})   #replace() takes dictionary only
df
```

```python
### Adding/Removing rows and columns from a dataFrame ###

df["first_name"]+" "+df["last_name"]    # firstname + space + lastname
```

```python
# Adding Column
df["full_name"]=df["first_name"]+" "+df["last_name"]
df
```

```python
# Remove columns
df.drop(columns=["full_name"],inplace=True)
```

```python
df
```

```python
# Adding Row
people2={"first_name":["Bruce","Robin"],"last_name":["Shrestha","kc"],"Email":["bruce@email.com","robin@email.com"]}
df2=pd.DataFrame(people2)
```

```python
df=pd.concat([df,df2],ignore_index=True)
```

```python
df
```

```python
# Remove Row
df.drop(index=7,inplace=True)
```

```python
df
```

```python
filt=(df["Email"]=="jane@email.com")
df.drop(index=df[filt].index, inplace=True)    # filtering and deleting entire row.
df
```

```python
## Sorting
df.sort_values(by="last_name",ascending=False,inplace=True)    # Descending
df
```

```python
df.loc[5,"last_name"]="potter"
df
```

```python
df.sort_values(by=["last_name","first_name"],ascending=False,inplace=True)
df
# first lastname bata sort garxa and if last name are same then goes to firstname. last name is in priority.
```
